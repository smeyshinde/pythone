```python
import matplotlib.pyplot as plt

#create a line plot
x=[1,2,3,4,5,6]
y=[4,7,8,9,5,4]
plt.plot(x,y)

#add lable and title
plt.title('simple Line Plot')
plt.xlabel('X axis')
plt.ylabel('y axis')
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,4,8,11,18]
plt.plot(x,y)

#create a scatter
plt.scatter(x,y)
#show plot
plt.show()
```
